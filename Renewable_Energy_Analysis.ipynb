# Business Problem 

Energy Crisis is a critical issue the world is facing, driven by increase in demand, geo-political problems and depletion of fossil fuels. Transitioning to Renewable Energy Sources is the key focus of governments worldwide to resolve energy crisis and ensure sustainability. In this case-study I am exploring the usage of global renewable energy sources - Solar, Wind, Hydro, Biomass and Geothermal.

# Data Source

<a id='summary'></a>
# Dataset Summary

The dataset contains global renewable energy adoption and usage details of 6 Regions spanning the years **2020** to **2024**, spread across 12 Columns with 1000 observations. It provides data to identify trends of **Renewable Energy Sources** such as **Solar, Wind, Hydro, Biomass and Geothermal** being utilized in **residential settings**. The dataset includes below listed variables/columns:  
1. Household ID -  unique identifier for each household
2. Region -   Geographical region where the household is located 
3. Country -  Country where the household is located
4. Energy_Source -  The renewable energy sources adopted by each household. (Solar, Hydro, Wind, Biomass and Geothermal)
5. Monthly_Usage_kWh -  Monthly energy consumption per household in kilowatt-hours
6. Year -  Year in which data is recorded
7. Household_Size -  The size/ number of members in each household
8. Income_Level -  The income class of each household (Low, High or Medium)
9. Urban/Rural -  Locality where the household is located
10. Adoption_Year -  The Year in which renewable energy source is adopted by that household
11. Subsidy_Received -  Whether or not the household has received goverment aid
12. Cost_Savings_USD -  The cost saved per month in USD by using renewable energy

## Renewable Energy Adoption Over Time

```python
# Understanding trend in renewable energy adoption over the years
adoption_trend = energy_usage.groupby(by='Adoption_Year', as_index=False)['Energy_Source'].count()

# The aggregated column no longer represents individual energy sources; it represents the count of energy sources adopted. Hence renaming for clarity
adoption_trend.rename(columns={'Energy_Source':'Total_Energy_Sources_Adopted'},inplace=True)

adoption_trend
```

Insights:
Above dataframe gives the total number of renewable energy sources adopted by households per year from 2010-2024. We can quickly identify that maximum number of adoptions happened in 2010. For quick/better understanding we can plot above data using a line graph. 

<a id='lineplot'></a>
### Visualization - Adoption Over Time

```python
# create a figure to plot the graph
plt.figure(figsize=(12,6))

# Generate line graph to show trends over time
sns.lineplot(data= adoption_trend, x= 'Adoption_Year', y='Total_Energy_Sources_Adopted', marker ='o')

# Adding Title 
plt.title('Renewable Energy Adoption Over Time')

plt.show()
```

**Insights**:
1. Overall, there is **decline** in renewable energy adoption at households from 2010 to 2024. Considering the factors like increase in demand and depletion of fossil fuels, an increase was expected in adoption.
2. The **highest adoption happened in 2010** (with 90 households adopted) followed by decline in 2011 and 2012.
3. From 2013 to 2015 there was a slight and stable improvement in adoption rate. (83,79 and 79 adoptions respectively).
4. 2017 showed a progress (with 80 adptions) similar to 2013 to 2015 range, after a decline in 2016(with 63 total adoptions).
5. From 2019 to 2021 again there was a slight yet stable increase seen in adoption.
6. **After 2021, in recent years, there is steep decline in adoption rate**.
7. The least adoption happened in 2024 as per the data, with 62% decrease from previous year. Month wise information is not available in dataset and hence its not clear if all 12 months of data were recorded or not. It might be because of incomplete data. Further analysis is needed to understand the real cause in this surprising decline.

<a id='consumption'></a>
## Energy Consumption per Energy Source

```python
# Finding the Total Consumption in kWh per Energy Sources
usage_per_source = energy_usage.groupby('Energy_Source',as_index=False)['Monthly_Usage_kWh'].sum()

# Calculating percentage contribution of each source to total monthly usage
usage_per_source['Usage %'] = (usage_per_source['Monthly_Usage_kWh']/(usage_per_source['Monthly_Usage_kWh'].sum())*100).round(1)

# sorting usage_per_sorce in desc order
usage_per_source= usage_per_source.sort_values(by='Usage %',ascending=False,ignore_index=True)
usage_per_source
```

<a id='consumptionviz'></a>
### Visualization of Consumption per Energy Source

```python
plt.figure(figsize=(8,5))
sns.set(style='whitegrid')
sns.barplot(data=usage_per_source, x='Energy_Source',y='Usage %')
plt.ylabel('Total Usage(%)')
plt.title('Monthly Energy Consumption per Source')
plt.show()
```

**Insights:**

1. Wind Energy contributes the highest to monthly energy usage at 23.4%, followed by Hydro with 20.1%. This indicates a dominant role of these two sources in total energy consumption.

2. Solar and Biomass represent 19.5% and 19.3% of total monthly usage, respectively, showing a relatively close distribution of consumption across different renewable energy sources.

3. Geothermal, although contributing the least in this list, still holds a significant share of 17.8%. Its contribution suggests that geothermal energy is still an important player in the energy mix, especially in areas where it is viable.

4. Trends between Adoption and Consumption: The energy consumption per source appears to roughly correlate with the adoption rates, suggesting that higher adoption generally leads to higher energy consumption. However, monthly usage can be influenced by a variety of factors such as:

       Household size: Larger households tend to consume more energy.

       Consumer mindset: Spending behavior on energy or the use of energy-efficient devices.

       Weather factors: Energy production from renewable sources like Solar and Wind can be significantly affected by weather conditions (e.g., cloudy days for solar or low wind days for wind energy).
